```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib as pyplot
import pickle
from matplotlib import style
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
import seaborn as sns
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('student-mat.csv', sep=";")
```

```python
df.head()
```

```python
# show corelation matrix
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.show()
```

```python
df=df[["G1","G2","G3","studytime","failures","absences"]]
predict="G3"
X= np.array(df.drop([predict], axis=1))
Y=np.array(df[predict])

best = 0
for _ in range(10000):
    x_train, x_test, y_train, y_test=train_test_split(X,Y,test_size=0.1)
    linear=linear_model.LinearRegression()
    linear.fit(x_train,y_train)
    accuracy=linear.score(x_test,y_test)
    if accuracy>best:
        with open("Studentmodel.pickle","wb") as f:
            pickle.dump(linear,f)
pickle_in = open("Studentmodel.pickle", "rb")
linear = pickle.load(pickle_in)
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.1)
accuracy = linear.score(x_test, y_test)
print("Accuracy:", accuracy)
```

```python
print("Co-efficient: \n", linear.coef_)
print("Intercept: \n", linear.intercept_)
predictions= linear.predict(x_test)

print("Actual Value\t\tGiven Input\t\tPredicted Value")
for x in range(len(predictions)):
    print("{}\t\t{}\t\t{:.5f}".format(y_test[x], x_test[x], predictions[x]))
```
